```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_games = pd.read_csv("partidos_cebollitas_9eada58c-fb57-4224-a3f5-6d9efc881c2e.csv")
df_games.head()
```

```python
#New  feature : difference goals (local - visitant)
df_games['diferencia_goles'] = df_games['goles_local'] - df_games['goles_visitante']
df_games[['goles_local','goles_visitante','diferencia_goles']].head()
```

```python
#New feaures: ratio shots to arco percentage lcoal
df_games['ratio_tiros_posesion_local'] = df_games['tiros_arco_local']/df_games['posesion_local (%)']
# see results
df_games[['tiros_arco_local','posesion_local (%)','ratio_tiros_posesion_local']].head()
```

```python
# histograms
sns.histplot(df_games['diferencia_goles'],kde=True,bins=15)
plt.title("Distribution differen goals")
plt.xlabel("Difference goals")
plt.ylabel("frecuency")
plt.show()
```

```python
#Correlation Analysis
corr_vars = ['goles_local','goles_visitante','diferencia_goles','ratio_tiros_posesion_local']
corr_matrix = df_games[corr_vars].corr()
corr_matrix
#Heatmap
sns.heatmap(corr_matrix , annot=True , cmap='coolwarm')
plt.show()
```
