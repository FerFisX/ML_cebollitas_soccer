```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load data
df_games = pd.read_csv("partidos_cebollitas_9eada58c-fb57-4224-a3f5-6d9efc881c2e.csv")
```

```python
df_games.head()
```

```python
df_games.describe()
```

```python
# Mean games by local and visitant
goals_local = df_games[df_games['equipo_local']== 'Cebollitas FC']['goles_local'].mean()
goals_visitant = df_games[df_games['equipo_visitante'] == 'Cebollitas FC']['goles_visitante'].mean()
print("Goals mean Cebollitas FC as local :" , goals_local)
print("Goals mean Cebollitas FC as visitant" , goals_visitant)
```

```python
# Histograms by goals (local VS Visitant)
fig, ax = plt.subplots(1,2,figsize=(12,4))
sns.histplot(df_games['goles_local'],ax=ax[0],kde=True,bins=10)
ax[0].set_title('Distribucion goles equipo local')
sns.histplot(df_games['goles_visitante'],ax=ax[1],kde=True,bins=10,color='b')
ax[1].set_title('Distribucion goles equipo visitante')
```

```python
#Boxplot Team local
sns.boxplot(x=df_games['goles_local'])
plt.title('Boxplot goal local team')
plt.show()
```

```python
#Scatterplot hold vs mark-goals
sns.scatterplot(x='posesion_local (%)',y='goles_local',data=df_games)
plt.title("Relation posesion vs mark_goals ('Local team')")
plt.show()
```

```python
# HEatmap key variables
plt.figure(figsize=(10,6))
sns.heatmap(df_games[['goles_local','goles_visitante','posesion_local (%)',
                      'posesion_visitante (%)','tiros_arco_local',
                      'tiros_arco_visitante']].corr(),annot=(1==1),cmap='coolwarm')
plt.title('Heatmap - correlation beetween variables')
plt.show()
```
